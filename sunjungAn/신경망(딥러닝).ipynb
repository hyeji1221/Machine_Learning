> 다층 퍼셉트론(MLP) : 분류와 회귀에 쓸 수 있음, 신경망이라고도 함

## 신경망 모델
+ 여러 단계를 거쳐 결정을 만들어냄
+ 가중치 합을 만드는 과정이 여러번 반복된다. (중간 단계 '은닉 유닛'을 추가할 수 있음)
+ 은닉 유닛의 가중치 합은 비선형 함수인 렐루나 하이퍼볼릭 탄젠트 활성화 함수를 적용함
**렐루** : 0이하를 자름
**하이퍼볼릭 탄젠트** : 크면 1로 수렴 작으면 -1로 수렴


## 신경망 튜닝


```python
import mglearn
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.neural_network import MLPClassifier
from sklearn.datasets import make_moons

X,y = make_moons(n_samples = 100, noise = 0.25, random_state = 3)

X_train, X_test, y_train, y_test = train_test_split(X,y,stratify = y, random_state = 42)

mlp = MLPClassifier(solver = 'lbfgs', random_state = 0, hidden_layer_sizes = [10]).fit(X_train, y_train) # solver -> 최적화 알고리즘
mglearn.plots.plot_2d_separator(mlp,X_train, fill = True, alpha = .3)
mglearn.discrete_scatter(X_train[:,0], X_train[:,1], y_train)
plt.xlabel("특성 0")
plt.ylabel("특성 1")

```

> 더 매끄러운 결정 경게는 은닉유닛, 은닉층을 추가하거나 tanh함수를 사용한다.

**은닉층 추가**

```python
mlp = MLPClassifier(solver = 'lbfgs', random_state = 0, hidden_layer_sizes = [10, 10]).fit(X_train, y_train) # solver -> 최적화 알고리즘
# 은닉 유닛이 10개 2층으로 구조
mglearn.plots.plot_2d_separator(mlp,X_train, fill = True, alpha = .3)
mglearn.discrete_scatter(X_train[:,0], X_train[:,1], y_train)
plt.xlabel("특성 0")
plt.ylabel("특성 1")
```

**tanh함수 적용**

```python
mlp = MLPClassifier(solver = 'lbfgs', random_state = 0, hidden_layer_sizes = [10, 10], activation = 'tanh').fit(X_train, y_train) # solver -> 최적화 알고리즘
# 은닉 유닛이 10개 2층으로 구조
mglearn.plots.plot_2d_separator(mlp,X_train, fill = True, alpha = .3)
mglearn.discrete_scatter(X_train[:,0], X_train[:,1], y_train)
plt.xlabel("특성 0")
plt.ylabel("특성 1")
```

**규제(alpha)값에 따른 변화**

```python

fig, axes = plt.subplots(2,4, figsize = (20,8))
for axx, n_hidden_nodes in zip(axes, [10,100]):
    for ax, alpha in zip(axx, [0.0001, 0.01, 0.1, 1]):
        mlp = MLPClassifier(solver = 'lbfgs', random_state = 0,
                           hidden_layer_sizes = [n_hidden_nodes, n_hidden_nodes], alpha = alpha)
        mlp.fit(X_train, y_train)
        mglearn.plots.plot_2d_separator(mlp, X_train, fill = True, alpha = .3, ax=ax)
        mglearn.discrete_scatter(X_train[:,0], X_train[:, 1], y_train, ax=ax)
        ax.set_title("n_hidden=[{},{}]\nalpha={:.4f}".format(n_hidden_nodes, n_hidden_nodes, alpha))
```

> 복잡도를 제어하는 방법 : 유닛, 은닉층의 개수, 규제
> 학습을 시작하기 전 가중치를 무작위로 설정 --> 학습에 영향을 줌

**가중치 초기화 값 변화**
+ 모델 학습 모양이 달라진 것을 확인할 수 있음

```python
fig, axes = plt.subplots(2,4, figsize = (20,8))
for i, ax in enumerate(axes.ravel()):
     mlp = MLPClassifier(solver = 'lbfgs', random_state = i, hidden_layer_sizes = [100,100])
     mlp.fit(X_train, y_train)
     mglearn.plots.plot_2d_separator(mlp, X_train, fill = True, alpha = .3, ax=ax)
     mglearn.discrete_scatter(X_train[:,0], X_train[:, 1], y_train, ax=ax)
```

```python
# 유방암 데이터셋에 MLP적용
from sklearn.datasets import load_breast_cancer

cancer = load_breast_cancer()
print("유방암 데이터의 특성별 최댓값 : \n", cancer.data.max(axis = 0))
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    cancer.data, cancer.target, random_state = 0)

mlp = MLPClassifier(random_state=42)
mlp.fit(X_train, y_train)

print("훈련 세트 정확도: {:.2f}".format(mlp.score(X_train ,y_train)))
print("테스트 세트 정확도 : {:.2f}".format(mlp.score(X_test, y_test)))
```

```python
# 데이터 스케일 변화 --> 평균은 0, 분산은 1로 조정
mean_on_train = X_train.mean(axis = 0)
std_on_train = X_train.std(axis = 0)

X_train_scaled = (X_train - mean_on_train)/std_on_train
X_test_scaled = (X_test - mean_on_train)/std_on_train

mlp = MLPClassifier(alpha = 1,max_iter = 1000, random_state = 0) # 반복횟수 증가, 가중치를 더 강하게 alpha 큰값으로
mlp.fit(X_train_scaled, y_train)

print("훈련 세트 정확도: {:.2f}".format(mlp.score(X_train_scaled ,y_train)))
print("테스트 세트 정확도 : {:.2f}".format(mlp.score(X_test_scaled, y_test)))
```

```python
plt.figure(figsize = (20,5))
plt.imshow(mlp.coefs_[0], interpolation='none', cmap ='viridis')
plt.yticks(range(30), cancer.feature_names)
plt.xlabel("은닉 유닛")
plt.ylabel("입력 특성")
plt.colorbar()

#입력과 은닉층 사이의 학습된 가중치
# 밝은 색은 큰 양수값, 어두운 색은 음수 값
```
