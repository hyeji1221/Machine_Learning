### k-최근접 이웃

+ 훈련 데이터셋을 저장하는 것이 모델을 만드는 과정
+ 훈련 데이터셋에서 가장 가까운 데이터 포인트 '최근접 이웃'을 찾는 것
+ 최근접 이웃을 임의의 k개로 선택할 수 있기 때문에 k-최근접 이웃이라고 한다. 
+ 여러개의 이웃을 선택할 때는 k-최근접 이웃 중 다수의 클래스가 레이블이 된다. 


```python
import mglearn
import matplotlib as plt

X,y = mglearn.datasets.make_forge()
mglearn.discrete_scatter(X[:, 0], X[:,-1], y)
mglearn.plots.plot_knn_classification(n_neighbors=1)
```

> **k-최근접 이웃 알고리즘 적용**

```python
import sklearn
from sklearn.model_selection import train_test_split
X,y = mglearn.datasets.make_forge() #데이터 만들기

X_train, X_test, y_train, y_test = train_test_split(X,y, random_state= 0) # 훈련세트 데이터세트 만들기

from sklearn.neighbors import KNeighborsClassifier
clf = KNeighborsClassifier(n_neighbors = 3) #이웃을 3개로하는 k-최근접 객체 생성
clf.fit(X_train, y_train)

print("테스트 세트 예측: ", clf.predict(X_test))
print("테스트 세트 정확도 : {:.2f}".format(clf.score(X_test, y_test)))
```

> **KNeighborsClassifier 분석** 

+ 클래스 0과 클래스 1로 지정한 영역으로 나뉘는 결정 경계를 볼 수 있다. 

```python
import matplotlib.pyplot as plt
import pandas as pd

fig, axes = plt.subplots(1,3, figsize=(10,3)) # 1행에 3개의 plot

for n_neighbors, ax in zip([1,3,9], axes):
    clf = KNeighborsClassifier(n_neighbors = n_neighbors).fit(X,y)
    mglearn.plots.plot_2d_separator(clf, X, fill = True, eps = 0.5, ax = ax, alpha = .4)
    mglearn.discrete_scatter(X[:, 0], X[:, 1], y, ax=ax)
    ax.set_title("{} 이웃".format(n_neighbors))
    ax.set_xlabel("특성 0")
    ax.set_ylabel("특성 1")
axes[0].legend(loc = 3)
```

+ 오른쪽으로 갈수록 경계선이 부드러워진다. --> 모델이 간단해진다.
+ 즉 k가 적을 수록 모델이 복잡해지고, 많아지면 간단해진다. 
+ 예를 들어 모든 데이터 포인터가 최근접 포인터라면 새로운 데이터의 예측값은 항상 같다. 


> **이웃의 수를 달리하여 훈련 세트와 테스트 세트 성능 평가**


```python
from sklearn.datasets import load_breast_cancer

cancer = load_breast_cancer()
X_train, X_test, y_train, y_test = train_test_split(
    cancer.data, cancer.target, stratify = cancer.target, random_state = 66)

training_accuracy = [] #정확도를 넣을 배열
test_accuracy = []

neighbors_settings = range(1,11)

for n_neighbors in neighbors_settings:
    clf = KNeighborsClassifier(n_neighbors= n_neighbors)
    clf.fit(X_train, y_train)
    training_accuracy.append(clf.score(X_train, y_train))
    test_accuracy.append(clf.score(X_test, y_test))
    
plt.plot(neighbors_settings, training_accuracy, label="훈련 정확도")
plt.plot(neighbors_settings, test_accuracy, label = "테스트 정확도")
plt.ylabel("정확도")
plt.xlabel("n_neighbors")
plt.legend()
```

> 테스트세트에 대해서 가장 좋은 정확도는 6개정도를 했을 때이다.
> 하지만 전체적으로 0.88정도의 정확도보다는 큰 것으로 보아 수긍할만하다.
